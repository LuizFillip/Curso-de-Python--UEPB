```python
from bibtexparser.bwriter import BibTexWriter
from bibtexparser.bibdatabase import BibDatabase
import re
import bibtexparser
```

# Open bibtex file

```python
#open bibtex file as string method
with open('011_Referencias.bib') as bibtex_file:
    bibtex_str = bibtex_file.read()

# load bibliografy with next function
bib_database = bibtexparser.loads(bibtex_str)
```

# Change some item into bibtex file


```python
db = BibDatabase()
#create data base of bibtex and append values with a empty list
db.entries = []

for num in range(len(bib_database.entries)):
    item = bib_database.entries[num]
    #sometimes, some errors may occur because the setting of items 
    #So, use "try" and "except"
    try: 
        #I need remove the char '{}' between names of firts authors
        item['author'] = item['author'].replace('{', '').replace('}', '')
        db.entries.append(item)
    except:
        pass

#function of writer of bibtex
writer = BibTexWriter()
filename = '011_Referencias.bib'

#write to file with filename declarante before
with open(filename, 'w') as bibfile:
    bibfile.write(writer.write(db))
```

```python
def find_indices_of(char, in_string):
    index = -1
    while True:
        index = in_string.find(char, index + 1)
        if index == -1:
            break
        yield index

def thesis_parser(bib):
    indexers = [i for i in find_indices_of('. ', bib)]

    without_authors = []
    authors = []

    for i in range(len(indexers) - 1):
        condition = indexers[i + 1] - indexers[i]
        if condition != 3:
            without_authors.append(bib[indexers[i]:].split('. '))
            authors.append(bib[:indexers[i]])
    
    try: 
        rest_infos = without_authors[0][4].split(' - ')      
        title = without_authors[0][1]
        pages = without_authors[0][2].replace(' p', '')
        notes = without_authors[0][3]
        school = rest_infos[1].split(',')[0]
        city = rest_infos[1].split(',')[1].strip()
        year = rest_infos[1].split(',')[2].replace('.', '').strip()

        author = authors[0]

        ID = author.split(',')[0] + year

        phdthesis = {
            'title' : title,
            'pages' : pages,
            'year': year,
            'school':school + ', ' + city,
            'ENTRYTYPE': 'phdthesis',
            'ID': ID,
            'author': author,
        }
        return phdthesis
    except:
        pass
    


```
