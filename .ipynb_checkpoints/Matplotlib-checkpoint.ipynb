# Matplotlib

Matplotlib  uma biblioteca abrangente para criar esttica, animada, e visualizaes interativas em Python. Matplotlib torna as coisas fceis coisas fceis e difceis possveis.



```python
import matplotlib.pyplot as plt 
import numpy as np 

x = np.arange(0, 100, 1)

# loop com varios periodos 

for period in np.arange(1, 10, 2):
    
    y = 4 * np.sin(np.pi * x / period)

    plt.plot(x, y, label = f'T = {period} s')
    plt.xlabel('time (s)')
    plt.ylabel('amplitude')
    plt.title('Funo seno')

    plt.legend()
```
