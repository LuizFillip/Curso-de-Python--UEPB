## O DataFrame

Um DataFrame do Pandas é uma estrutura bidimensional semelhante a uma tabela em Python, onde os dados são organizados em linhas e colunas. É uma das ferramentas mais usadas para lidar com dados e facilita a organização, análise e manipulação de dados. Ele pode armazenar diferentes tipos de dados, como números, texto e datas em suas colunas. As principais partes de um DataFrame são:

- Dados: valores reais na tabela.
- Linhas: rótulos que identificam cada linha.
- Colunas: rótulos que definem cada categoria de dados.

## Criação de um DataFrame


```python
import numpy as np

arr = np.random.rand(1000, 3)

df = pd.DataFrame(arr, columns = ['age', 'fare', 'height'])

df.index = pd.date_range('2025-09-12', freq = '1D', periods = arr.shape[0])

ds = df.loc[df.index.day == 6, ['height']]

ds['volume'] = np.pi * 3/2 * ds['height']**3

ds 
```

# Xarray 

Matrizes multidimensionais (também conhecidas como N-dimensionais, ND) (às vezes chamadas de "tensores") são uma parte essencial da ciência computacional. Eles são encontrados em uma ampla gama de campos, incluindo física, astronomia, geociência, bioinformática, engenharia, finanças e aprendizado profundo. Em Python, o NumPy fornece a estrutura de dados fundamental e a API para trabalhando com matrizes ND brutas. No entanto, os conjuntos de dados do mundo real geralmente são mais do que apenas números brutos; eles têm rótulos que codificam informações sobre como os valores da matriz são mapeados para locais no espaço, tempo, etc.





```python
import xarray as xr
import numpy as np 
import pandas as pd 


```

```python
#Dados do satelite GOES

infile = 'E:\\database\\goes\\2018\\12\\S10635346_201812010000.nc'

ds = xr.open_dataset(infile)

ds['data'] = ds['Band1']  / 100 - 273.13

ds 
```

```python
#Ler vários arquivos

```

```python
import matplotlib.pyplot as plt
import cartopy.feature as cf
import cartopy.crs as ccrs
import numpy as np 

fig, ax = plt.subplots(
    figsize = (5, 5), 
    dpi = 100,
    subplot_kw = 
        {'projection': ccrs.PlateCarree()}
)

ax.add_feature(cf.COASTLINE)

ylimits = [-60, 30]
xlimits = [-90, -30]
ax.set_extent(
    [xlimits[0], xlimits[-1], 
     ylimits[0], ylimits[-1]], 
        crs = ccrs.PlateCarree())

ax.set_yticks(
    np.arange(ylimits[0], ylimits[-1], 15)
)
ax.set_xticks(
    np.arange(xlimits[0], xlimits[-1], 30)
)

img = ax.contourf(
    ds.lon.values,
    ds.lat.values,
    ds.data.values, 
    cmap = 'jet'
)

plt.colorbar(img)
```
