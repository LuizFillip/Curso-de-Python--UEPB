## O que  um "Array"?

Em programao de computadores, um array (**matriz**)  uma estrutura para armazenar e recuperar dados. Costumamos falar sobre um array como se fosse uma grade no espao, com cada clula que armazena um elemento dos dados. Por exemplo, se cada elemento do dados fossem um nmero, poderamos visualizar um array "unidimensional" como um lista:



Um array bidimensional seria como uma tabela:


 
Um array tridimensional seria como um conjunto de mesas, talvez empilhadas como se fossem impressos em pginas separadas. No NumPy, essa ideia  generalizado para um nmero arbitrrio de dimenses e, portanto, o classe de matriz  chamada : representa um "array N-dimensional".ndarray

A maioria das arrays tem algumas restries. Por exemplo:

- Todos os elementos da matriz devem ser do mesmo tipo de dados.

- Depois de criado, o tamanho total da matriz no pode ser alterado.

- A forma deve ser "retangular", no "irregular"; por exemplo, cada linha de um array bidimensional deve ter o mesmo nmero de colunas.

## Incializando Arrays

O NumPy fornece vrias funes integradas para gerar matrizes com propriedades especficas.

- np.zeros(): Cria um array preenchido com zeros.
- np.ones(): Cria um array preenchido com uns.
- np.full(): Cria um array preenchido com um valor especificado.
- np.arange(): Cria uma matriz com valores uniformemente espaados dentro de um determinado intervalo.
- np.linspace(): Cria uma matriz com valores espaados uniformemente em um intervalo especificado.

## Gerao de nmeros aleatrios

O NumPy fornece funes para criar matrizes preenchidas com nmeros aleatrios.

- np.random.rand(): Cria uma matriz de forma especificada e a preenche com valores aleatrios amostrados de uma distribuio uniforme em [0, 1).
- np.random.randn(): Cria uma matriz de forma especificada e a preenche com valores aleatrios amostrados de uma distribuio normal padro.
- np.random.randint(): Cria uma matriz de forma especificada e a preenche com inteiros aleatrios dentro de um determinado intervalo.

```python
np.random.rand(100).sum()
```

### Indexao na matriz NumPy

A diviso e a indexao bsicas so usadas para acessar um elemento especfico ou intervalo de elementos de uma matriz NumPy.

O fatiamento e a indexao bsicos retornam apenas a exibio da matriz.

Considere a sintaxe x[obj] onde "x"  o array e "obj"  o ndice. O objeto slice  o ndice no caso de fatiamento bsico.

O fatiamento bsico ocorre quando obj :

- Um objeto slice que tem o formato start: stop: step
- Um nmero inteiro
- Ou uma tupla de objetos de fatia e inteiros
- Todas as matrizes geradas pelo fatiamento bsico so sempre 'vistas' da matriz original.

```python
b = np.arange(0, 9, 1).reshape(3, 3)

b[1:, 1:]
```

```python
a = np.arange(20) 
print("\n Array is:\n ",a) 
print("\n a[15]=",a[15])
# a[start:stop:step]
print("\n a[-8:17:1] = ",a[-8:17:1]) 
print("\n a[10:] = ",a[10:]) 

 
a   
```
