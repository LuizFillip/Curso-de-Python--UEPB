### Dicionrio

Um dicionrio em Python  uma coleo de valores de dados, usada para armazenar valores de dados como um mapa, ao contrrio de outros tipos de dados Python, um dicionrio contm um par chave (**key**): valor. A chave-valor  fornecida no dicionrio para torn-la mais otimizada. Cada par de valores-chave em um dicionrio  separado por dois pontos : , enquanto cada chave  separada por uma 'vrgula'.

Criar um dicionrio em Python

Os valores em um dicionrio podem ser de qualquer tipo de dados e podem ser duplicados, enquanto as chaves no podem ser repetidas e devem ser imutveis. O dicionrio tambm pode ser criado pela funo embutida dict().

```python
#Exemplo-Idades das pessoas da sala
idades = {'ryan': 21, 'edu': 36, 'luiz': 30, 'louri': 65, 'gio': 14}

# obtendo na forma de lista as chaves e os valores 
keys = list(idades.keys())
values = list(idades.values())

# Forma de trabalhar simultaneamente com cada item do dicionrio 
for key, value in idades.items():
    
    if value < 18:
        print(f'{key}  uma criana')
    elif (value >= 20) and (value <= 30):
        print(f'{key}  um jovem')
    elif (value > 30) and (value < 60):
        print(f'{key}  um adulto')
    else:
        print(f'{key}  um idoso')
        
```

## Tratamento de excees do Python

O tratamento de excees do Python lida com erros que ocorrem durante a execuo de um programa. O tratamento de excees permite responder ao erro, em vez de travar o programa em execuo. Ele permite que voc capture e gerencie erros, tornando seu cdigo mais robusto e fcil de usar. Vejamos um exemplo:

Manipulando uma exceo simples em Python

- O tratamento de excees ajuda a evitar falhas devido a erros. Aqui est um exemplo bsico que demonstra como capturar uma exceo e trat-la normalmente:

```python
# Diviso por zero
n = 2
for n in range(-2, 10, 2):
    try: 
        result = 10 / n
    except:
        result = 0

    print(f'10 dividido por {n}  igual a {result}') 
```

### Diferena entre exceo e erro

- **Erros** so problemas srios que um programa no deve tentar resolver. Geralmente so problemas na lgica ou configurao do cdigo e precisam ser corrigidos pelo programador. Os exemplos incluem erros de sintaxe e erros de memria.

- As **excees** so menos graves do que os erros e podem ser tratadas pelo programa. Eles ocorrem devido a situaes como entrada invlida, arquivos ausentes ou problemas de rede.

## Classes (class)

Uma classe em Python  um modelo definido pelo usurio para criar objetos. Ele agrupa dados e funes, facilitando o gerenciamento e o uso deles. Quando criamos uma nova classe, definimos um novo tipo de objeto. Podemos ento criar vrias instncias desse tipo de objeto.

Por que precisamos de classes e objetos

- Suporta programao orientada a objetos usando modelos reutilizveis (classes) e modelos do mundo real (objetos).
- Promove a reutilizao de cdigo e o design modular com mtodos e atributos organizados.
- Simplifica programas complexos agrupando dados e comportamentos relacionados.


```python
class Students:
    
    names = ['ryan', 'edu']
    ages = [21, 36]
    
    def __init__(self, name, age):
        self.name = name 
        self.age = age 
        
        
    def msg(self):
        return f'{self.name} tem {self.age} anos'
    
    def eh_o_que(self):
        key = self.name
        value = self.age
        if value < 18:
            print(f'{key}  uma criana')
        elif (value >= 20) and (value <= 30):
            print(f'{key}  um jovem')
        elif (value > 30) and (value < 60):
            print(f'{key}  um adulto')
        else:
            print(f'{key}  um idoso')
            
            
clas = Students('Anderson', 34)

clas.eh_o_que()
```

### Mdulo datetime 

Em Python, data e hora no so tipos integrados, mas so tratados usando o mdulo datetime integrado . Este mdulo oferece aulas para trabalhar eficientemente com datas, horrios e intervalos, fornecendo muitos mtodos teis. Date e DateTime so objetos, portanto, manipul-los significa trabalhar com objetos, no com strings simples ou carimbos de data/hora.

Por que precisamos do mdulo Datetime?
Ajuda a trabalhar com datas e horas em aplicativos do mundo real, como agendamento ou registro em log.
Permite o clculo fcil de diferenas entre duas datas ou horas.
Suporta formatao e anlise de strings de data/hora para sadas fceis de usar.
til para eventos, arquivos ou entradas de dados com carimbo de data/hora.
Essencial para lidar com fusos horrios, duraes e operaes baseadas em calendrio.
Classes comumente usadas no mdulo datetime
Classe

| Python      | Descrio                                                                               |
| ----------- | ------------------------------------------------------------------------------------------------------- |
| `date`      | Representa uma **data** (ano, ms, dia) no calendrio gregoriano.                                       |
| `time`      | Representa uma **hora** independente de qualquer data (hora, minuto, segundo, microssegundo, `tzinfo`). |
| `datetime`  | Combina **data e hora** (ano, ms, dia, hora, minuto, segundo, microssegundo, `tzinfo`).                |
| `timedelta` | Representa o **delta de tempo**, ou seja, a diferena entre duas datas ou horas.                        |
| `tzinfo`    | Classe base abstrata para informaes de **fuso horrio**.                                              |
| `timezone`  | Classe de **fuso horrio fixo por deslocamento** (subclasse de `tzinfo`), introduzida no Python 3.2.    |
