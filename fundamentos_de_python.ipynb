### Tipos de dados numéricos

Os números do Python representam dados que têm um valor numérico. Um valor numérico pode ser um número inteiro, um número flutuante ou até mesmo um número complexo. Esses valores são definidos como classes int, float e complex.

- Inteiros - o valor é representado pela classe int. Ele contém números inteiros positivos ou negativos (sem frações ou decimais). Não há limite para o comprimento que um valor inteiro pode ter.
- Float - o valor é representado pela classe float. É um número real com uma representação de ponto flutuante. É especificado por um ponto decimal. Opcionalmente, o caractere e ou E seguido por um inteiro positivo ou negativo pode ser anexado para especificar a notação científica.
- Números complexos - É representado por uma classe complexa. É especificado como (parte real) + (parte imaginária)j. Por exemplo - 2+3j

### Dicionário

Um dicionário em Python é uma coleção de valores de dados, usada para armazenar valores de dados como um mapa, ao contrário de outros tipos de dados Python, um dicionário contém um par chave (**key**): valor. A chave-valor é fornecida no dicionário para torná-la mais otimizada. Cada par de valores-chave em um dicionário é separado por dois pontos : , enquanto cada chave é separada por uma 'vírgula'.

Criar um dicionário em Python

Os valores em um dicionário podem ser de qualquer tipo de dados e podem ser duplicados, enquanto as chaves não podem ser repetidas e devem ser imutáveis. O dicionário também pode ser criado pela função embutida dict().

```python
#Exemplo-Idades das pessoas da sala
idades = {'ryan': 21, 'edu': 36, 'luiz': 30, 'louri': 65, 'gio': 14}

# obtendo na forma de lista as chaves e os valores 
keys = list(idades.keys())
values = list(idades.values())

# Forma de trabalhar simultaneamente com cada item do dicionário 
for key, value in idades.items():
    
    if value < 18:
        print(f'{key} é uma criança')
    elif (value >= 20) and (value <= 30):
        print(f'{key} é um jovem')
    elif (value > 30) and (value < 60):
        print(f'{key} é um adulto')
    else:
        print(f'{key} é um idoso')
        
```

## Tratamento de exceções do Python

O tratamento de exceções do Python lida com erros que ocorrem durante a execução de um programa. O tratamento de exceções permite responder ao erro, em vez de travar o programa em execução. Ele permite que você capture e gerencie erros, tornando seu código mais robusto e fácil de usar. Vejamos um exemplo:

Manipulando uma exceção simples em Python

- O tratamento de exceções ajuda a evitar falhas devido a erros. Aqui está um exemplo básico que demonstra como capturar uma exceção e tratá-la normalmente:

```python
# Divisão por zero
n = 2
for n in range(-2, 10, 2):
    try: 
        result = 10 / n
    except:
        result = 0

    print(f'10 dividido por {n} é igual a {result}') 
```

### Diferença entre exceção e erro

- **Erros** são problemas sérios que um programa não deve tentar resolver. Geralmente são problemas na lógica ou configuração do código e precisam ser corrigidos pelo programador. Os exemplos incluem erros de sintaxe e erros de memória.

- As **exceções** são menos graves do que os erros e podem ser tratadas pelo programa. Eles ocorrem devido a situações como entrada inválida, arquivos ausentes ou problemas de rede.

## Classes (class)

Uma classe em Python é um modelo definido pelo usuário para criar objetos. Ele agrupa dados e funções, facilitando o gerenciamento e o uso deles. Quando criamos uma nova classe, definimos um novo tipo de objeto. Podemos então criar várias instâncias desse tipo de objeto.

Por que precisamos de classes e objetos

- Suporta programação orientada a objetos usando modelos reutilizáveis (classes) e modelos do mundo real (objetos).
- Promove a reutilização de código e o design modular com métodos e atributos organizados.
- Simplifica programas complexos agrupando dados e comportamentos relacionados.


```python
class Students:
    
    names = ['ryan', 'edu']
    ages = [21, 36]
    
    def __init__(self, name, age):
        self.name = name 
        self.age = age 
        
        
    def msg(self):
        return f'{self.name} tem {self.age} anos'
    
    def eh_o_que(self):
        key = self.name
        value = self.age
        if value < 18:
            print(f'{key} é uma criança')
        elif (value >= 20) and (value <= 30):
            print(f'{key} é um jovem')
        elif (value > 30) and (value < 60):
            print(f'{key} é um adulto')
        else:
            print(f'{key} é um idoso')
            
            
clas = Students('Anderson', 34)

clas.eh_o_que()
```

### Módulo datetime 

Em Python, data e hora não são tipos integrados, mas são tratados usando o módulo datetime integrado . Este módulo oferece aulas para trabalhar eficientemente com datas, horários e intervalos, fornecendo muitos métodos úteis. Date e DateTime são objetos, portanto, manipulá-los significa trabalhar com objetos, não com strings simples ou carimbos de data/hora.

Por que precisamos do módulo Datetime?
Ajuda a trabalhar com datas e horas em aplicativos do mundo real, como agendamento ou registro em log.
Permite o cálculo fácil de diferenças entre duas datas ou horas.
Suporta formatação e análise de strings de data/hora para saídas fáceis de usar.
Útil para eventos, arquivos ou entradas de dados com carimbo de data/hora.
Essencial para lidar com fusos horários, durações e operações baseadas em calendário.
Classes comumente usadas no módulo datetime
Classe

| Python      | Descrição                                                                               |
| ----------- | ------------------------------------------------------------------------------------------------------- |
| `date`      | Representa uma **data** (ano, mês, dia) no calendário gregoriano.                                       |
| `time`      | Representa uma **hora** independente de qualquer data (hora, minuto, segundo, microssegundo, `tzinfo`). |
| `datetime`  | Combina **data e hora** (ano, mês, dia, hora, minuto, segundo, microssegundo, `tzinfo`).                |
| `timedelta` | Representa o **delta de tempo**, ou seja, a diferença entre duas datas ou horas.                        |
| `tzinfo`    | Classe base abstrata para informações de **fuso horário**.                                              |
| `timezone`  | Classe de **fuso horário fixo por deslocamento** (subclasse de `tzinfo`), introduzida no Python 3.2.    |
